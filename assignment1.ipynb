```python
def start_game():
    print("Welcome to the Adventure Game!")
    print("You find yourself at a crossroad in a dark forest.")
    first_choice()

def first_choice():
    print("\nDo you want to go LEFT or RIGHT?")
    choice = input("Enter 'left' or 'right': ").strip().lower()

    if choice == "left":
        left_path()
    elif choice == "right":
        right_path()
    else:
        print("Invalid choice. Please choose 'left' or 'right'.")
        first_choice()

def left_path():
    print("\nYou head left and arrive at a wide river.")
    print("There's a bridge, but it looks old and shaky.")
    print("Alternatively, you could try to swim across.")
    second_choice_left()

def second_choice_left():
    print("\nDo you want to SWIM across or LOOK for a bridge?")
    choice = input("Enter 'swim' or 'look': ").strip().lower()

    if choice == "swim":
        print("\nYou decide to swim across the river.")
        print("Halfway through, a river monster appears and drags you underwater.")
        print("Game Over!")
    elif choice == "look":
        print("\nYou decide to look for a bridge.")
        print("After a short walk, you find a sturdy bridge and cross safely.")
        print("On the other side, you discover a hidden treasure. You Win!")
    else:
        print("Invalid choice. Please choose 'swim' or 'look'.")
        second_choice_left()

def right_path():
    print("\nYou head right and come across a dark cave.")
    print("The cave looks inviting but mysterious.")
    print("You can choose to enter the cave or continue walking through the forest.")
    second_choice_right()

def second_choice_right():
    print("\nDo you want to ENTER the cave or CONTINUE walking?")
    choice = input("Enter 'enter' or 'continue': ").strip().lower()

    if choice == "enter":
        print("\nYou bravely enter the cave.")
        print("Inside, you find a sleeping dragon guarding a pile of gold.")
        print("The dragon wakes up and, seeing you, breathes fire.")
        print("Game Over!")
    elif choice == "continue":
        print("\nYou decide not to enter the cave and keep walking through the forest.")
        print("After hours of walking, you get lost and can't find your way out.")
        print("Game Over!")
    else:
        print("Invalid choice. Please choose 'enter' or 'continue'.")
        second_choice_right()

if __name__ == "__main__":
    start_game()
```
